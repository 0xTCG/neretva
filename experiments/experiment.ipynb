Use the following command for benchmark the tool on CYP and KIR genes:
#### KIR

parallel --bar -j2 'python neretva/kir.py --input {} > results_validation/{/.}.log 2>&1' ::: data/hprc/*.cram

#### CYP
```bash

PREFIX=$1

mkdir -p "$PREFIX"

for gene in CYP2C8 CYP2C9 CYP2C19 CYP2D6; do
    mkdir -p "${PREFIX}/${gene}"
    ls /cyp-data/*.wgs.cram | \
    /parallel -j6 \
    "id=\$(basename {} .wgs.cram); /time -v python neretva/cyp.py ${gene} {} > ${PREFIX}/${gene}/\${id}.log 2>&1"
done
```

## Result summarization and evaluation procedure
### Parsing the raw results
- To summarize the KIR results into Google sheet compatible format, use: 
```bash
python experiments/scripts/make_kir_table.py --truth experiments/ground-truth/hprc_gt_0801.csv --results experiments/results/results_KIR_HPRC > experiments/sheets/HPRC_sheet.csv
```
### Marking and Evaluation
We use the following scripts for metric evaluation, with the graded results in [summarized sheet](experiments/sheets/KIR_CYP_marked.csv).
- To get all the CYP results summarized in one table for Google sheet compatible format, use 
```bash
python experiments/scripts/make_cyp_table.py
```
- To get the metrics for KIR, use the following scripts:
```bash
python experiments/scripts/mark_sample_level.py experiments/sheets/KIR_CYP_marked.csv
python experiments/scripts/mark_allele_level.py experiments/sheets/KIR_CYP_marked.csv
```
- To get the metrics for CYP, use the following scripts:
```bash
python experiments/scripts/mark_sample_level_CYP.py experiments/sheets/KIR_CYP_marked.csv
python experiments/scripts/mark_allele_level_CYP.py experiments/sheets/KIR_CYP_marked.csv
```
