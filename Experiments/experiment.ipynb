## Result summarization and evaluation procedure
- To summarize the KIR results into Google sheet compatible format, use: 
```bash
python Scripts/make_kir_table.py --truth Ground_truth/hprc_gt_0801.csv --results ../results/results_KIR_HPRC > Sheets/HPRC_sheet.csv
```
- To get all the CYP results summarized in one table for Google sheet compatible format, use 
```bash
python Scripts/make_cyp_table.py
```
- To get the metrics for KIR, use the following scripts:
```bash
python Scripts/mark_sample_level.py Sheets/KIR_CYP_marked.csv
python Scripts/mark_allele_level.py Sheets/KIR_CYP_marked.csv
```
- To get the metrics for CYP, use the following scripts:
```bash
python Scripts/mark_sample_level_CYP.py Sheets/KIR_CYP_marked.csv
python Scripts/mark_allele_level_CYP.py Sheets/KIR_CYP_marked.csv
```
- Find the summarized and marked results of each of the KIR and CYP tools at [Collected results](./Sheets/KIR_CYP_marked.csv)
- To get the per-gene breakdown results for each of the tools evaluated, use:
```bash
python Scripts/mark_gene.py Sheets/KIR_CYP_marked.csv > Sheets/KIR_per_gene_break_down.tex
```
