## Result summarization and evaluation procedure
- To summarize the KIR results into Google sheet compatible format, use: `python Scripts/make_kir_table.py --truth Ground_truth/hprc_gt_0801.csv --results ../results/results_KIR_HPRC > Sheets/HPRC_sheet.csv`
- Find the summarized and marked results of each of the KIR and CYP tools at [`Collected results`](./Sheets/KIR_CYP_marked.csv)
- To get the per - gene breakdown results for each of the tools evaluated, use: `python Scripts/mark_gene.py Sheets/KIR_CYP_marked.csv > Sheets/KIR_per_gene_break_down.tex`.
